```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score


# Load the house prediction dataset
data = pd.read_csv('./housing.csv')  # Replace 'house_prediction_dataset.csv' with the actual dataset file name

# Split the dataset into features (X) and target variable (y)
X_single = data[['Size']]  # Single-variable case
X_multi = data[['Size', 'Bedrooms']]  # Multiple-variable case
y = data['Price']

# Create linear regression models for both single-variable and multiple-variable cases
reg_single = LinearRegression()
reg_multi = LinearRegression()

# Fit the models using the training data
reg_single.fit(X_single, y)
reg_multi.fit(X_multi, y)

# Make predictions using the trained models
y_pred_single = reg_single.predict(X_single)
y_pred_multi = reg_multi.predict(X_multi)

# Calculate evaluation metrics for single-variable case
mse_single = mean_squared_error(y, y_pred_single)
r2_single = r2_score(y, y_pred_single)

# Calculate evaluation metrics for multiple-variable case
mse_multi = mean_squared_error(y, y_pred_multi)
r2_multi = r2_score(y, y_pred_multi)

# Print evaluation metrics
print("Single-Variable Linear Regression:")
print("Mean Squared Error:", mse_single)
print("R^2 Score:", r2_single)

print("\nMultiple-Variable Linear Regression:")
print("Mean Squared Error:", mse_multi)
print("R^2 Score:", r2_multi)

# Plot the actual and predicted prices for the single-variable case
plt.figure(figsize=(8, 6))
plt.scatter(X_single, y, color='b', label='Actual')
plt.plot(X_single, y_pred_single, color='r', label='Predicted')
plt.xlabel('Size')
plt.ylabel('Price')
plt.title('Single-Variable Linear Regression')
plt.legend()
plt.show()

# Plot the actual and predicted prices for the multiple-variable case
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_multi['Size'], X_multi['Bedrooms'], y, c='b', label='Actual')
ax.scatter(X_multi['Size'], X_multi['Bedrooms'], y_pred_multi, c='r', label='Predicted')
ax.set_xlabel('Size')
ax.set_ylabel('Bedrooms')
ax.set_zlabel('Price')
ax.set_title('Multiple-Variable Linear Regression')
ax.legend()
plt.show()
```
